#In SeniorCitizen column the values were 0(False) or 1(True), we are Defining a function for converting O or 1 to Yes or No for good understanding

```python
def conv(value):
    if(value==1):
        return "Yes"
    else:
        return "No"

df["SeniorCitizen"]= df["SeniorCitizen"].apply(conv)
```

```python
plt.figure(figsize = (4,4))
ax=sns.countplot(x="Churn",data = df )
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb=df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels=gb.index,autopct = "%1.2f%%")
plt.title("Percentage Of Churned Customers")
plt.show()
```

The Pei-Chart we see above shows that 26.54% Customers churned out.
Now Let find out what are the reasons behind it.

```python
plt.figure(figsize=(5,4,))
aw=sns.countplot(x="gender",data = df , hue="Churn")
aw.bar_label(aw.containers[1])
aw.bar_label(aw.containers[0])
plt.title("Counting Based on Gender those Churned Out/Not")
plt.show()
```

```python
plt.figure(figsize=(5,4,))
aw=sns.countplot(x="SeniorCitizen",data= df)
aw.bar_label(aw.containers[0])
plt.title("Number of Senior Citizen..")
plt.show()
```

```python
data = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='Count')
total_counts = data.groupby('SeniorCitizen')['Count'].transform('sum')
data['Percentage'] = data['Count'] / total_counts * 100

# Pivot the data for easier plotting
pivot_data = data.pivot(index='SeniorCitizen', columns='Churn', values='Percentage').fillna(0)

# Step 2: Plot the stacked bar chart
fig, ax = plt.subplots(figsize=(5, 4))

categories = pivot_data.index  # SeniorCitizen categories
bottoms = [0] * len(categories)

for churn_status in pivot_data.columns:
    percentages = pivot_data[churn_status]
    ax.bar(categories, percentages, bottom=bottoms, label=churn_status)
    # Add percentage labels on bars
    for i, (percent, bottom) in enumerate(zip(percentages, bottoms)):
        if percent > 0:  # Skip if percentage is 0
            ax.text(i, bottom + percent / 2, f'{percent:.1f}%', ha='center', va='center', color='white', fontsize=10)
    bottoms = [b + p for b, p in zip(bottoms, percentages)]

# Step 3: Add labels, legend, and title
ax.set_xlabel("Senior Citizen")
ax.set_ylabel("Percentage")
ax.set_title("Churn by Senior Citizen (as % of total)")
ax.legend(title="Churn", loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()

# Show the plot
plt.show()
```

Comparative a greater percentage of people from Senior Citizen category are churned out.

```python
# Columns to plot
columns_to_plot = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the subplot grid
n_cols = 3  # Number of columns in the grid
n_rows = (len(columns_to_plot) + n_cols - 1) // n_cols  # Calculate rows needed
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 14), constrained_layout=True)

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Loop through the columns and create a plot for each
for idx, column in enumerate(columns_to_plot):
    es=sns.countplot(x=column, data=df, ax=axes[idx],hue="Churn")
    es.bar_label(es.containers[0])
    es.bar_label(es.containers[1])
    axes[idx].set_title(f"Churn by {column}")
    
    axes[idx].set_xlabel(column)
    axes[idx].set_ylabel("Count")

# Remove any extra/empty subplots
for idx in range(len(columns_to_plot), len(axes)):
    fig.delaxes(axes[idx])

# Show the plots
plt.show()
```

Customers without services like OnlineSecurity, TechSupport, or DeviceProtection tend to have a higher churn rate, while those with these services generally churn less. Additionally, customers with Fiber optic internet service show a notably higher churn rate compared to other internet service types. These insights highlight potential areas for improving customer retention.
